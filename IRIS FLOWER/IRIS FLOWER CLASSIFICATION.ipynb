# SPLIT THE DATASET

```python
x= data.drop('species', axis=1)
y = data['species']
print("Shape of X = ",x.shape)
print("Shape of Y = ",y.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=50)

print("Shape of X_train = ", x_train.shape)
print("Shape of X_test = ", x_test.shape)
print("Shape of y_train = ", y_train.shape)
print("Shape of y_test = ", y_test.shape)
```

# MODEL FITTING

```python
DecisionTree = DecisionTreeClassifier()

DecisionTree.fit(x_train, y_train)
DecisionTree_score = DecisionTree.score(x_test, y_test)
DecisionTree_score
```

```python
KNN = KNeighborsClassifier()

KNN.fit(x_train, y_train)
KNN_score = KNN.score(x_test, y_test)
KNN_score
```

```python
svc = SVC(kernel='linear')

svc.fit(x_train, y_train)
svc_score = svc.score(x_test, y_test)
svc_score
```

```python
RandomForest = RandomForestClassifier()

RandomForest.fit(x_train, y_train)
RandomForest_score = RandomForest.score(x_test, y_test)
RandomForest_score
```

```python
gaussion = GaussianNB()

gaussion.fit(x_train, y_train)
gaussion_score = gaussion.score(x_test, y_test)
gaussion_score
```

```python
multinomial = MultinomialNB()

multinomial.fit(x_train, y_train)
multinomial_score = multinomial = multinomial.score(x_test, y_test)
multinomial_score
```

```python
bernoulli = BernoulliNB()

bernoulli.fit(x_train, y_train)
bernoulli_score = bernoulli.score(x_test, y_test)
bernoulli_score
```

# PREDICTION MODEL AND ASSOCIATED SCORES

```python
score_table = pd.DataFrame({"Test":['Decision Tree', 'KNN', 'Support Vector', 'Random Forest', 'Gaussion NB', 
                                    'Multinomial NB', 'Bernoulli NB'], 
              "Score":[DecisionTree_score, KNN_score, svc_score, RandomForest_score, gaussion_score, multinomial_score, 
                       bernoulli_score]})
score_table
```
